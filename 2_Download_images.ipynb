# Content based

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# TF-IDF vectorizer
vectorizer = TfidfVectorizer(stop_words='english', max_features=5000)
tfidf_matrix = vectorizer.fit_transform(df['text'])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel

vectorizer = TfidfVectorizer(dtype=np.float32)
tfidf_matrix = vectorizer.fit_transform(df['text'].fillna(''))

cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
# Map indices to product IDs
indices = pd.Series(df.index, index=df['ProdID'])

def content(prod_id, top_n=5):

    idx = indices[prod_id]  # Get index for the product ID
    sim_scores = list(enumerate(cosine_sim[idx]))  # Get cosine similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)[1:top_n+1]  # Sort and slice top_n results
    product_indices = [i[0] for i in sim_scores]  # Extract indices of similar products
    return df.iloc[product_indices][['ProdID', 'Name']]  # Return relevant columns

# Example of dynamically passing top_n when calling the function
prod_id = "B0CGPN9VNK"  # Example product ID
top_n = 10  # Dynamically set top_n to 10
content_rec = content(prod_id, top_n)
print(content_rec)
```

# Item Collaboratory recommendation


```python

# Function to get recommendations based on similar star ratings
def collab(prod_id, top_n=5):
    """
    Get top_n similar products to a given product based on star rating similarity.

    Parameters:
    - prod_id: Product ID for which recommendations are to be generated.
    - top_n: The number of recommendations to return (default is 5).
    """
    # Get the rating (star) of the product
    prod_rating = df[df['ProdID'] == prod_id]['stars'].values[0]

    # Find products with similar ratings (we can define a threshold or just pick the same star rating)
    similar_products = df[df['stars'] == prod_rating]
    
    # If there are too few similar products, expand to products with close ratings
    if len(similar_products) < top_n:
        similar_products = df[abs(df['stars'] - prod_rating) <= 0.5]  # Consider products within 0.5 stars

    # Sort by star rating (optional: could sort by name, ID, etc. for variety)
    similar_products = similar_products.sort_values(by='stars', ascending=False)
    
    # Get the top N similar products
    top_similar_products = similar_products.head(top_n)
    
    return top_similar_products[['ProdID', 'Name', 'stars']]

# Example usage
prod_id = "B0CGPN9VNK"  # Example product ID
top_n = 10  # Number of recommendations
collab_rec = collab(prod_id, top_n)
print(collab_rec)
```

```python
import pickle 

# Load the extracted image features from the pickle file
with open(r'C:\Users\VAGDEVI\image_features_all.pkl', 'rb') as f:
    image_features = pickle.load(f)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
# Create a matrix of features for cosine similarity calculation
feature_matrix = np.array(list(image_features.values()))

# Calculate the cosine similarity between all products
cosine_sim = cosine_similarity(feature_matrix)

# We will use a Pandas DataFrame to make the result easier to interpret
cosine_sim_df = pd.DataFrame(cosine_sim, index=df['ProdID'], columns=df['ProdID'])
```

```python

def image_based(prod_id, top_n=5):
    """
    Get top_n similar products to a given product based on image similarity.
    
    Parameters:
    - prod_id: Product ID for which recommendations are to be generated.
    - top_n: The number of recommendations to return (default is 5).
    
    Returns:
    - A DataFrame with the top_n similar products based on image similarity.
    """
    # Get similarity scores for the product (prod_id) in descending order
    sim_scores = cosine_sim_df[prod_id].sort_values(ascending=False)
    
    # Exclude the first product (itself)
    sim_scores = sim_scores[1:top_n+1]
    
    # Get product IDs of the top N similar products
    top_product_ids = sim_scores.index
    
    # Return product details (e.g., ProdID, Name, and similarity score)
    recommendations = df[df['ProdID'].isin(top_product_ids)][['ProdID', 'Name']]
    recommendations['Similarity'] = sim_scores.values
    
    return recommendations

```

```python
# Example usage
prod_id = "B0C85L8MZM"  # Example product ID (adjust this to a valid product ID)
top_n = 5  # Number of recommendations
image_rec = image_based(prod_id, top_n)
print(image_rec)
```
